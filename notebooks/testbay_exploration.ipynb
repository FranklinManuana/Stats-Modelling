## Make function to go through json file to find networks in tokyo

```python
'''Create a function to parse through information to pull
info for tokyo '''
def tokyo_networks(res):
    network_list = []
    for i in range(len(res)):
        if res[i]['location']['city'].lower() == 'tokyo':
            network_list.append(res[i])
    return network_list
print(tokyo_networks(res))

```

```python
def tokyo_networks(n2):
    network_list = []
    for i in range(len(n2)):
        if 'tokyo' in n2[i]['location']['city'].lower():
            network_list.append(n2[i])
    return network_list
values_tokyo = tokyo_networks(n2)
print(values_tokyo)
```

## Obtain FOURSQUARE API information

```python
# import libraries
import requests
import pandas as pd
import os
import json
stations_test
df
```

```python
# make request to connect to FOURSQUARE API and store 1 location
api_key = os.environ["FOURSQUARE_API_KEY"]
headers = {"Accept": "application/json"}
headers['Authorization'] = api_key


def test_call(stations_test):
    url ="https://api.foursquare.com/v3/places/search?ll="
    for i in range(1):
        latitude  = stations_test[i]['lat']
        longitude = stations_test[i]['lon']
        ll = str(latitude) +','+ str(longitude)
        full_url = url + ll
        response_fs = requests.get(full_url,headers=headers)
    return response_fs.json()
test_result = test_call(stations_test)

```

```python
df1 = pd.json_normalize(test_result['results'])
df2 = pd.json_normalize(test_result['results'])
df1['station'] = 1
df2['station'] = 2
df_set = [df1, df2]
df_set = pd.concat(df_set)
df_set.groupby("station").agg({'distance': 'min', 'fsq_id': 'count'})
```

```python
# Create JSON file for foursquare test call
with open('fs_tokyo_test_call.json', 'w') as f:
    json.dump(test_result, f)
```

```python
# store created JSON file into a variable for exploration
with open('fs_tokyo_test_call.json') as f:
    tokyo_fs_test = json.load(f)
print(tokyo_fs_test)
```
